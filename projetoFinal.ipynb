```python
# Bibiotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Lendo a base de dados
Base_Dados = pd.read_csv('./materiais/Dados_Performance_Estudantes.csv')

# Verificando os primeiros registros
Base_Dados.head()
```

```python
# Tamanho da Imagem
fig, ax = plt.subplots( figsize=(18, 15) )

# Lista com as paletas de cores
Paleta_Cores = sns.color_palette('husl', 20)

# Cor de Fundo da Imagem
Cor_Fundo = '#f5f5f5'
ax.set_facecolor( Cor_Fundo )
fig.set_facecolor( Cor_Fundo )

# Titulo
plt.suptitle('Imersão no framework SEABORN \n Ciência de Dados para Finanças e Economia', 
             fontsize=22, color=Paleta_Cores[1], fontweight=600 )

# Estilo do grafico
plt.style.use('ggplot')

#################
### gráfico 1 ###
#################
plt.subplot(3, 2, 1)

# Titulo
plt.title('Análise 1 - Nota Matematica x Redação', loc='left', fontsize=15)
# plot
sns.scatterplot( data=Base_Dados, x='math score', y='reading score' )

#################
### gráfico 2 ###
#################
plt.subplot(3, 2, 2)

# Titulo
plt.title('Análise 2 - Distrbuição das Notas', loc='left', fontsize=15)
# plot
Ordenacao = ['group A', 'group B', 'group C', 'group D', 'group E']
sns.boxplot( data=Base_Dados, y='math score', x='race/ethnicity', hue='gender', 
            palette='bone_r', linewidth=2.5, order=Ordenacao )

#################
### gráfico 3 ###
#################
plt.subplot(3, 2, 3)
# Titulo
plt.title('Análise 3 - Correlação das Notas', loc='left', fontsize=15)
# Plot
Correlacao = Base_Dados.corr()
sns.heatmap( Correlacao, annot=True, cmap='YlGnBu')

#################
### gráfico 4 ###
#################
plt.subplot(3, 2, 4)
# Titulo
plt.title('Análise 4 - Quantidade Notas', loc='left', fontsize=15)
sns.histplot( data=Base_Dados, x='math score', hue='gender', element='step')

#################
### gráfico 5 ###
#################
plt.subplot(3, 2, 5)
# Titulo
plt.title('Análise 5 - Quantidade dos Grupos', loc='left', fontsize=15)
sns.barplot( data=Base_Dados, y='math score', x='race/ethnicity', color='salmon', saturation=.5 )

#################
### gráfico 6 ###
#################
plt.subplot(3, 2, 6)
# Titulo
plt.title('Análise 6 - Notas curso preparatório', loc='left', fontsize=15)
sns.lineplot( data=Base_Dados, x='math score', y='reading score', 
             hue='gender', style='test preparation course')

# Ajustar os gráficos
plt.subplots_adjust( hspace=0.35 )

# Rodape
Rodape = '''
Esse relatorio foi elaborado no curso "Ciência de dados para Finanças e Economia" \n
by: 
@Alessa Santos
@Odemir Depieri Jr
'''

# Incluindo no relatorio o rodape
fig.text( 0.5, -0.03, Rodape, ha='center', va='bottom', size=12, color='#938ca1' );
```
