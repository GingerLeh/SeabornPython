Ferramentas do Seaborn
1. Gráficos relacionais: apresentam relação entre dois atributos
2. Gráficos categóricos: mostram fronteiras de categorias nos dados
3. Gráficos de distribuição: Mostram a distribuição de um conjunto de dados
4. Gráfico de regressão: relacionados a modelos de regressão
5. Gráficos de matriz: visualização em forma de matrizes

```python
# Bibiotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Lendo a base de dados
Base_Dados = pd.read_csv('./materiais/Dados_Performance_Estudantes.csv')

# Verificando os primeiros registros
Base_Dados.head()
```

### Gráfico de dispersão
**Scatterplot**

Documentação: https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
plt.figure( figsize=(12, 6) )
plt.title('Análise da nota de matemática x redação')
sns.scatterplot( data=Base_Dados, x='math score', y='writing score');
```

```python
# HUE
plt.figure( figsize=(12, 6) )
plt.title('Análise da nota de matemática x redação')
sns.scatterplot( data=Base_Dados, x='math score', y='writing score', hue='gender');
```

```python
# HUE
plt.figure( figsize=(12, 6) )
plt.title('Análise da nota de matemática x redação')
sns.scatterplot( data=Base_Dados, x='math score', y='writing score', hue='gender', style='test preparation course', palette='deep');
```

```python
Base_Dados.head()
```

```python
# Transformação de categorico para numerico

# Biblioteca - Função do Sklearn
from sklearn.preprocessing import LabelEncoder 

# Atribuindo a função
Funcao_Conversao = LabelEncoder()
```

```python
# Criando a coluna e atribuindo os valores
Base_Dados['Grupo'] = Funcao_Conversao.fit_transform(Base_Dados['race/ethnicity'])

# Verificando
Base_Dados.head()
```

```python
# HUE
plt.figure( figsize=(12, 6) )
plt.title('Análise do Grupo x Nota')
sns.scatterplot( data=Base_Dados, x='math score', y='Grupo',hue='gender', style='test preparation course');
```

```python
# Usando marcadores como dicionario

# Definindo marcadores
Marcadores = {'female' :'s', 'male' : 'X'}

# Plotagem
plt.figure( figsize=(12, 6) )
plt.title('Notas matemática vs Grupo')
sns.scatterplot( data=Base_Dados, x='math score', y='Grupo',hue='gender', style='gender', markers=Marcadores, palette='deep');
```

### Gráfico de Calor
**Heatmap**

https://seaborn.pydata.org/generated/seaborn.heatmap.html

![heatmap](./materiais/fotos/heatmap.png)

```python
# Calcular correlação
Correlacao = Base_Dados.corr()
```

```python
# Tamanho
plt.figure( figsize=(10, 6))

# Titulo
plt.title('Analise Correlação')

# Plot
sns.heatmap( Correlacao );
```

```python
# Tamanho
plt.figure( figsize=(10, 6))

# Titulo
plt.title('Analise Correlação')

# Plot [ANOOT + cbar]
sns.heatmap( Correlacao, annot=True, cbar=False);
```

```python
# Tamanho
plt.figure( figsize=(10, 6))

# Titulo
plt.title('Analise Correlação')

# Plot [ANOOT + cbar + linewidth]
sns.heatmap( Correlacao, annot=True, cbar=False, linewidth=0.9);
```

```python
# Tamanho
plt.figure( figsize=(10, 6))

# Titulo
plt.title('Analise Correlação')

# Plot [cmap]
sns.heatmap( Correlacao, cmap='YlGnBu');
```

```python
# Tamanho
plt.figure( figsize=(10, 6))

# Titulo
plt.title('Analise Correlação')

# Plot [cmap + xticklabels + yticklabels]
sns.heatmap( Correlacao, cmap='YlGnBu', xticklabels=False,yticklabels=False );
```

```python
# Gerando analise da media por nivel educacional dos pais
Analise_Media = Base_Dados.groupby( by=['parental level of education']).mean().drop(columns=['Grupo'])
```

```python
# Tamanho
plt.figure( figsize=(10, 6))

# Titulo
plt.title('Análise das medias por nivel educacional dos pais')

# Plot [cmap + xticklabels + yticklabels]
sns.heatmap( Analise_Media, cmap='BuPu', annot=True);
```
